```python
desserts = ('Lava Cake', 'Mousse', 'Chocolate Cake', 'Ice Cream', 'Truffles', 'Brownie', 'Chocolate Chip Cookie', 'Chocolate Pudding', 'Souffle', 
        'Chocolate Cheesecake','Chocolate Chips','Fudge', 'Mochi')
sales = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Dessert': desserts,
    'Sale': sales, 
})
df.sort_values(by='Sale', inplace=True)

x_coords = np.arange(len(df))
colors = ['#0000FF' for _ in range(len(df))]
colors[-11] = '#FF0000'
colors[-12] = '#FF0000'
colors[-13] = '#FF0000'
plt.bar(x_coords, df['Sale'], tick_label=df['Dessert'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Sales Count')
plt.title('Dessert Sales')
plt.show()
```
